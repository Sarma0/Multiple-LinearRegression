__Importing the libraries__

```python
import pandas as pd
import numpy as np
```

__Importing the dataset__

```python
dataset = pd.read_csv(r'C:\Users\Toshiba\Documents\datasets\Fish Market dataset.csv')
dataset
```

```python
x = dataset.iloc[:,2:].values
y = dataset.iloc[:,1].values
print(x)
print(y)
```

__Spliting data into Training set and Testing set__

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train,)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
a = dataset.columns[2:]
b = regressor.coef_
c = pd.DataFrame(b,index = a,columns = ['Coefficients'])
c
```

```python
regressor.score(x_train,y_train)
```

__Predicting the Test set results__

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
regressor.score(x_test,y_test)
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df1 = df.head(30) 
df1
```
